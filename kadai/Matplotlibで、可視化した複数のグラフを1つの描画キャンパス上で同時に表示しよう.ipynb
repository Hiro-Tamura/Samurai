折れ線グラフ、sin関数、ヒストグラム、散布図のデータを可視化して、グラフや図を1つの描画キャンパス上にまとめて表示してください。

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, axes = plt.subplots(2, 3, tight_layout=True)

line_data_num = 5
rng = np.random.default_rng()
#lineX = np.arange(line_data_num)
lineY = rng.random(line_data_num)
axes[0,0].plot(lineY)

sinX = np.linspace(0, 10, 100)
sinY = 2 + 2 * np.sin(2 * sinX)
axes[0,1].plot(sinX, sinY)

histX = np.random.normal(15, 5, 2000)
axes[0,2].hist(histX)

np.random.seed(3)
scatX = 4 + np.random.normal(0, 2, 60)
scatY = 4 + np.random.normal(0, 2, len(scatX))

# プロットする
axes[1,0].scatter(scatX, scatY)

plt.show()
```
